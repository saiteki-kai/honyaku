```python
import typing

from pathlib import Path

import pandas as pd

from datasets import Dataset
```

```python
# dataset_name = "saiteki-kai__BeaverTails-Evaluation-it"
dataset_name = "saiteki-kai__BeaverTails-it"

metrics = [
    "google__metricx-24-hybrid-xxl-v2p6",
    "Unbabel__wmt23-cometkiwi-da-xxl",
    "Unbabel__XCOMET-XXL",
]

models = [
    "haoranxu__X-ALMA-13B-Group2",
    "Unbabel__TowerInstruct-Mistral-7B-v0.2",
    "CohereForAI__aya-23-35B",
    "facebook__nllb-200-3.3B",
    "facebook__nllb-moe-54B",
    "LLaMAX__LLaMAX3-8B-Alpaca",
]

splits = ["330k_train", "330k_test"]
# splits = ["test"]

for metric in metrics:
    df_results = pd.DataFrame()
    for model in models:
        df = pd.DataFrame()
        for split in splits:
            filepath = Path(f"../outputs/{dataset_name.split('__')[-1]}/scores/{metric}/{model}/{split}/scores.parquet")

            if not filepath.exists():
                print(filepath)
                continue

            ds = Dataset.from_parquet(str(filepath))
            ds = typing.cast(Dataset, ds)
            df_split = typing.cast(pd.DataFrame, ds.to_pandas())

            df = pd.concat([df, df_split])

        df["model"] = model
        df["model"] = pd.Categorical(df["model"], models)
        df = df.rename(columns={col: col.split("_")[0] for col in df.columns})

        df_results = pd.concat([df_results, df])

    print(metric)
    p = 100 if "metricx" not in metric else 1

    # display(df_results.groupby("model", observed=True).agg(lambda x: f"{x.mean()*p:.2f} Â± {x.std()*p:.2f}"))
    display(df_results.groupby("model", observed=True).agg(lambda x: f"{x.mean() * p:.2f}"))  # noqa: B023
    print("\n")
```
