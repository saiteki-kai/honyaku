# Dataset

This notebook analyzes token statistics of the dataset and visualizes the longest prompts and responses.

```python
import typing

import pandas as pd

from datasets import Dataset, load_dataset
from vllm import LLM

from .utils.stats import dataset_statistics


pd.set_option("display.max_colwidth", None)
pd.set_option("display.width", 150)
```

```python
NEW_MAX_TOKENS = 600
SEPARATOR = "-" * 80

dataset_name = "PKU-Alignment/BeaverTails"
model_name = "haoranxu/X-ALMA-13B-Group2"

splits = ["330k_test", "330k_train"]
fields = ["prompt", "response"]
```

```python
tokenizer = LLM(model_name, enforce_eager=True).get_tokenizer()
```

```python
df = {}

for split in splits:
    print(f"{dataset_name} - {split}\n")

    dataset = load_dataset(dataset_name, split=split)
    dataset = typing.cast(Dataset, dataset)

    df[split] = dataset_statistics(dataset, tokenizer, fields)  # type: ignore  # noqa: PGH003

    for field in fields:
        df[split][field] = df[split][field].astype("string")

    df[split]["category"] = df[split]["category"].astype("string")
    print()

dataset_df = pd.concat([df["330k_train"], df["330k_test"]])
```

### Longest Responses

```python
dataset_df.sort_values(["response_token_len", "prompt_token_len"], ascending=False)
```

### Longest Prompts

```python
dataset_df.sort_values(["prompt_token_len", "response_token_len"], ascending=False)
```
